```python
string_file = "../datasets/pairs/human_train.pos.tsv"
df_string  = pd.read_csv(string_file, sep = "\t", header = None)
str_test   = "../datasets/pairs/human_test.pos.tsv"
df_test    = pd.read_csv(str_test, sep = "\t", header = None)
df_string  = pd.concat([df_string, df_test])
```

```python
df_string
```

```python
string_prots_all = set(df_string[0]).union(set(df_string[1]))
string_prots_all = list(string_prots_all)
len(string_prots_all)
```

```python
df_symbol = pd.read_csv("ensp_prot.tab", sep = "\t")
st_ensp   = df_symbol["yourlist:M202202204ABAA9BC7178C81CEBC9459510EDDEA3422E7DQ"]
st_sym    = df_symbol["Gene names  (primary )"]
ensp_sym  = {ensp:sym for ensp, sym in zip(st_ensp, st_sym) if sym != None}
ensp_sym, len(ensp_sym)
```

```python
from tqdm import tqdm
estring = {}
for i in tqdm(range(len(st_sym))):
    st = st_sym[i]
    try:
        estring[st] = mg.query(st, fields = "entrezgene")['hits'][0]["entrezgene"]
    except:
        print(f"No match found for {st}")
```

```python
ensp_entrez = {f"9606.{key}": estring[st] for key, st in ensp_sym.items() if st in estring}
ensp_keys   = set(ensp_entrez.keys())
string_filtered = df_string[0].isin(ensp_keys) & df_string[1].isin(ensp_keys)
df_string_filtered = df_string.drop(df_string.index[~string_filtered]).reset_index(drop = True)
```

```python
df_string_filtered
```

```python
df_string_filt = df_string_filtered.replace({0:ensp_entrez, 1:ensp_entrez})
```

```python
df_string_fin = df_string_filt.drop([2], axis = 1)
```

```python
df_string_fin.to_csv("human.STRING.entrez.tsv", sep = "\t", header = None, index = None)
```
